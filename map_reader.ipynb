```python
from pylab import *
```

```python
offset_to_seconds = 20 * 10
Rh_values = [1, 10, 40, 160]
i_values = [10, 30, 50, 70, 90, 110, 130, 150, 170]
```

```python
import numpy as np

for Rh in Rh_values:
    for i in i_values:
        # Define just the base name of the file here
        base_name = f'lc_Rh{Rh}_i{i}.npz'
        
        # Construct the full path for loading
        load_path = f'a-1o2/lightcurves/{base_name}'
        
        lc = np.load(load_path) # Also, ensure you use np.load
        
        # 1. Haal de data uit het geladen 'lc' object
        time = lc['offsets'] * offset_to_seconds
        flux = lc['fluxes'][:, 0]  
        
        # 2. Flux error
        flux_error = flux * 1E-3
        
        # 3. Samenvoegen
        data_to_save = np.column_stack((time, flux, flux_error))
        
        # 4. Opslaan (using base_name so the .txt file isn't nested in weird folders)
        save_path = f'../Pioran.jl/examples/ultranest/data/a-1o2/{base_name}.txt'
        np.savetxt(save_path, data_to_save, fmt='%f', comments='')
        
        print(f"Bestand {base_name}.txt is succesvol aangemaakt!")
```
